```python
#download dataset
od.download("https://www.kaggle.com/datasets/prathamsaraf1389/spotify-global-top-50-daily-update")
```

```python
#load data
df = pd.read_csv("spotify-global-top-50-daily-update/playlist.csv")
```

### Data Visualization

```python
#Most popular Track

popular_track = df.groupby('track_name')['track_popularity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,8))
plt.xticks(rotation=90)
plt.bar(popular_track.index, popular_track)
plt.title('Most Popular Song by Popularity')
plt.xlabel('Song Title')
plt.ylabel('Popularity')
```

```python
#Most popular artists

popular_track = df.groupby('name_of_artists')['track_popularity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,8))
plt.xticks(rotation=90)
plt.bar(popular_track.index, popular_track)
plt.title('Most Popular Artists')
plt.xlabel('Artists')
plt.ylabel('Popularity')
```

```python
#Most popular Track

popular_track = df.groupby('album_name')['track_popularity'].sum().sort_values(ascending=False).head(10)
plt.figure(figsize=(10,8))
plt.xticks(rotation=90)
plt.bar(popular_track.index, popular_track)
plt.title('Most Popular Album by Popularity')
plt.xlabel('Album Title')
plt.ylabel('Popularity')
```
